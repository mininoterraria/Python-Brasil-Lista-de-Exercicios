```python
#17) Faça um programa que calcule o fatorial de um número inteiro fornecido pelo usuário. Ex.: 5!=5.4.3.2.1=120

numero = int(input("Insira um número positivo inteiro: "))

fatorial = 1 #Variável que será o ponto de partida da fatoração, pois carrega o número primordial.
exibir_numeros = '' #Variável que exibirá os números da fatoração.

numeros_fatoracao = [] #Lista em que ficará todos os valores da fatoração do número primordial.

if(numero <= 0):
  print("Inválido, somente números positivos inteiros são aceitos.")
else:

  #Estrutura que fará o cálculo de fatoração, é encerrado quando o multiplicador é igual a 0, para não afetar o resultado final.
  for i in range(numero,0, -1):

      numeros_fatoracao.append(i)
      fatorial *= i

  #Estrutura para exibir formalmente o cálculo da fatoração
  exibir_numeros = list(map(str, numeros_fatoracao)) #A função map percorre cada elemento da lista numeros_fatoracao e converte um por um em string, por fim, transforma uma nova lista com a função list.
  exibir_numeros = '.'.join(exibir_numeros) #Substitui cada separador antigo de cada elemento pelo ponto.
  print(f'{numero}! = {exibir_numeros} = {fatorial}' )
```

```python
#35) Encontrar números primos é uma tarefa difícil. Faça um programa que gera uma lista dos números primos existentes entre 1 e um número inteiro informado pelo usuário.

while(True):

  numeros_primos = [] #Variável onde ficará armazenado os números primos para exibição.

  divisores = 0 #Variável que contabilizará a quantidade de divisores de um número, para saber se é primo ou não.

  numero = int(input("Insira um número inteiro: "))

  #Validação do número inputado.
  if(numero < 0):
    print("Inválido, insira somente valores positivos.")
    continue

  #Estrutura de laços que pega um número e vai dividindo por seus decrescentes.
  for i in range(numero,0,-1):
    divisores = 0

    for j in range(i,0,-1):
      if(i % j == 0):
        divisores += 1

    if(divisores == 2): #Se o número tiver apenas 2 divisores nesse intervalo, será dado como primo.
      numeros_primos.append(i)

  numeros_primos.reverse() #Função que inverte a ordem dos elementos da lista.

  if(len(numeros_primos) != 0):
    print("Números primos nesse intervalo:",*numeros_primos)
  else:
    print("Não há números primos nesse intervalo.")
```

```python
"""
#41) Faça um programa que receba o valor de uma dívida e mostre uma tabela com os seguintes dados: valor da dívida, valor dos juros,
quantidade de parcelas e valor da parcela.

Os juros e a quantidade de parcelas seguem a tabela abaixo:
Quantidade de Parcelas  % de Juros sobre o valor inicial da dívida
1       0
3       10
6       15
9       20
12      25

Exemplo de saída do programa:
Valor da Dívida Valor dos Juros Quantidade de Parcelas  Valor da Parcela
R$ 1.000,00     0               1                       R$  1.000,00
R$ 1.100,00     100             3                       R$    366,00
R$ 1.150,00     150             6                       R$    191,67
"""

#Dicionário que armazenará os dados referente à divida.
divida = {
    'valor_divida' : [],
    'juros': [],
    'qtd_parcelas' : [],
    'valor_parcela' : [],
}

while(True):
  quantidade_divida = int(input("Quantas dívidas você possui?: "))

  #Validação da quantidade de dívidas, todo o código será encerrado quando todas as dívidas forem preenchidas.
  if(quantidade_divida <= 0):
    print("Inválido, ao menos uma dívida.")
    continue
  break

iteracao_atual = 0 #Variável utilizada para iterar sobre elementos das listas.

divida_atual = 1 #Variável que exibe a dívida atual.

while(divida_atual <= quantidade_divida):

  while(True):

    valor_divida = float(input(f"Insira o valor da dívida {divida_atual}: "))

    #Validação do valor da dívda.
    if(valor_divida <= 0):
      print("Inválido, insira apenas valores positivos maiores que 0.")
      continue

    break

  while(True):

    qtd_parcelas = int(input(f"Insira a quantidade de parcelas da dívida {divida_atual}: "))

    #Validação da quantidade de parcelas
    if(qtd_parcelas <= 0):
      print("Inválido, insira apenas valores positivos maiores que 0.")
      continue

    elif(qtd_parcelas > 12):
      print("Inválido, 12 é a quantidade máxima de parcelamento.")
      continue

    elif(qtd_parcelas not in [1,3,6,9,12]):
      print("Inválido, o parcelamento só pode ser feito em 1,3,6,9,12 vezes.")
      continue

    #Aplicação dos juros de acordo com a quantidade de parcelas.
    if(qtd_parcelas == 1):
      divida['juros'].append(0 * valor_divida)

    elif(qtd_parcelas == 3):
      divida['juros'].append(0.10 * valor_divida)

    elif(qtd_parcelas == 6):
      divida['juros'].append(0.15 * valor_divida)

    elif(qtd_parcelas == 9):
      divida['juros'].append(0.20 * valor_divida)

    elif(qtd_parcelas == 12):
      divida['juros'].append(0.25 * valor_divida)

    #Soma o valor dos juros ao da dívida e armazena na lista.
    divida['valor_divida'].append(valor_divida + divida['juros'][iteracao_atual])

    #Armazena a quantidade de parcelas em sua lista.
    divida['qtd_parcelas'].append(qtd_parcelas)

    #Faz o cálculo do parcelamento e armazena em sua lista.
    divida['valor_parcela'].append(divida['valor_divida'][iteracao_atual] / divida['qtd_parcelas'][iteracao_atual])

    iteracao_atual += 1

    divida_atual += 1

    break

iteracao_atual = 0 #Essa mesma variável será utilizada para iterar na exibição de dados na tabela abaixo.

print(f"{'Valor da Dívida':<18} {'Valor dos Juros':<18} {'Quantidade de Parcelas':<22} {'Valor da Parcela':<18}")

# Loop para imprimir os valores com formatação de largura.
for i in range(1,divida_atual):

    print(f"R$ {divida['valor_divida'][iteracao_atual]:<17.2f} R$ {divida['juros'][iteracao_atual]:<17.2f} {divida['qtd_parcelas'][iteracao_atual]:<21} R$ {divida['valor_parcela'][iteracao_atual]:<17.2f}")

    iteracao_atual += 1
```

```python
"""
#45) Desenvolver um programa para verificar a nota do aluno em uma prova com 10 questões, o programa deve perguntar
ao aluno a resposta de cada questão e ao final comparar com o gabarito da prova e assim calcular o total de acertos
 e a nota (atribuir 1 ponto por resposta certa). Após cada aluno utilizar o sistema deve ser feita uma pergunta se
 outro aluno vai utilizar o sistema. Após todos os alunos terem respondido informar:
Maior e Menor Acerto;
Total de Alunos que utilizaram o sistema;
A Média das Notas da Turma.
Gabarito da Prova:
01 - A
02 - B
03 - C
04 - D
05 - E
06 - E
07 - D
08 - C
09 - B
10 - A
Após concluir isto você poderia incrementar o programa permitindo que o professor digite o gabarito da prova antes dos alunos usarem o programa.
"""

#Dicionário onde ficará armazenado as informações da avaliação e dos alunos.
avaliacao = {
    'quantidade_alunos': 1,
    'notas_adicionadas': 0,
    'questao_atual': 1,
    'gabarito': [],

    'dados_aluno' : {
        1: {
            'id': 1,
            'respostas_prova': [],
            'pontuacao_aluno': 0
        }
    }

}

avaliacao['questao_atual'] = 1 #Variável que iterará sobre cada questão para ser atribuído valor, seja do gabarito ou da prova
#de um aluno.

i = 1 #Variável que iterará sobre o perfil de cada aluno que está no dicionário.

pergunta = '' #Variável que servirá para encerrar o laço principal que faz o registro de dados dos alunos.

#Estrutura de laço para registro das alternativas do gabarito.
while(avaliacao['questao_atual'] <= 10):

    resposta_gabarito = input(f"Professor(a), insira a resposta da questão {avaliacao['questao_atual']} do gabarito: ").upper().strip()

    #Validação da resposta do(a) professor(a).
    if(resposta_gabarito not in ['A','B','C','D','E']):
      print("Inválido, insira apenas uma das letras válidas como resposta.")
      continue

    avaliacao['gabarito'].append(resposta_gabarito)

    avaliacao['questao_atual'] += 1

#Estrutura de laço para registro dos dados dos alunos.
avaliacao['questao_atual'] = 1

while(pergunta != 'N'):

  while(avaliacao['questao_atual'] <= 10):

    resposta_aluno = input(f"Aluno {avaliacao['dados_aluno'][i]['id']}, insira a resposta da questão {avaliacao['questao_atual']}: ").upper().strip()

    #Validação da resposta do(a) aluno(a).
    if(resposta_aluno not in ['A','B','C','D','E']):
      print("Inválido, insira apenas uma das letras válidas como resposta.")
      continue

    avaliacao['dados_aluno'][i]['respostas_prova'].append(resposta_aluno)

    avaliacao['questao_atual'] += 1

  while(True):

      pergunta = input("Outro(a) aluno(a) usará o sistema? (S) / (N): ").upper().strip()

      #Validação da pergunta
      if(pergunta == 'S'):
        i += 1

        avaliacao['questao_atual'] = 1

        avaliacao['quantidade_alunos'] += 1

        avaliacao['dados_aluno'][i] = {
            'id': i,
            'respostas_prova': [],
            'pontuacao_aluno': 0
          }

        break

      elif(pergunta == 'N'):
        break

      else:
        print("Inválido, insira apenas S(sim) ou N(não) como resposta.")
        continue

#Estrutura de laços para somar as notas de cada aluno.
for j in avaliacao['dados_aluno']:

  iterar_respostas = 0

  for k in avaliacao['gabarito']:

    if(avaliacao['dados_aluno'][j]['respostas_prova'][iterar_respostas] == k):
      avaliacao['dados_aluno'][j]['pontuacao_aluno'] += 1

    iterar_respostas += 1

#Estrutura de laços para definir a maior e menor nota.
menor_nota = float('inf') #Variável onde ficará armazenada a menor nota.

maior_nota = 0 #Variável onde ficará armazenada a maior nota.

for j in avaliacao['dados_aluno']:

  if(avaliacao['dados_aluno'][j]['pontuacao_aluno'] > maior_nota):
    maior_nota = avaliacao['dados_aluno'][j]['pontuacao_aluno']

  if(avaliacao['dados_aluno'][j]['pontuacao_aluno'] < menor_nota):
    menor_nota = avaliacao['dados_aluno'][j]['pontuacao_aluno']

#Estrutura para exibir a média de notas da turma.
media = 0 #Variável que exibirá a média.

for j in avaliacao['dados_aluno']:

    avaliacao['notas_adicionadas'] += avaliacao['dados_aluno'][j]['pontuacao_aluno']

media = (avaliacao['notas_adicionadas'] / avaliacao['quantidade_alunos'])

print(f"Menor acerto registrado: {menor_nota}")

print(f"Maior acerto registrado: {maior_nota}")

print(f"Quantidade de alunos que utilizaram o sistema: {avaliacao['quantidade_alunos']}")

print(f"A média de notas da turma é: {media:.2f}")
```

```python
"""
#46) Em uma competição de salto em distância cada atleta tem direito a cinco saltos. No final da série de saltos de cada atleta,
o melhor e o pior resultados são eliminados. O seu resultado fica sendo a média dos três valores restantes.
Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe a média dos
saltos conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média). Faça uso de uma lista para
armazenar os saltos. Os saltos são informados na ordem da execução, portanto não são ordenados. O programa deve ser encerrado quando
não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo:
Atleta: Rodrigo Curvêllo

Primeiro Salto: 6.5 m
Segundo Salto: 6.1 m
Terceiro Salto: 6.2 m
Quarto Salto: 5.4 m
Quinto Salto: 5.3 m

Melhor salto:  6.5 m
Pior salto: 5.3 m
Média dos demais saltos: 5.9 m

Resultado final:
Rodrigo Curvêllo: 5.9 m
"""

#Dicionário que armazena os dados dos competidores.
competicao = {
    1: {
        'id': 1,
        'nome': '',
        'saltos': [],
    }
}

pergunta = '' #Variável que servirá para finalizar o laço de cadastro de nomes e saltos dos competidores.

competidor_atual = 1 #Variável que iterará sobre cada competidor para fazer seu devido registro no dicionário.

validar_nome = True #Variável que servirá como parte da validação do nome, evitando pôr números.

while(pergunta != 'N'):

  while(True):
    validar_nome = True

    nome = input(f"Insira o nome do competidor {competidor_atual}: ").capitalize()

    #Validação do nome.
    for i in nome:
      if(i.isdigit()):
        print("Inválido, nome não pode ter números.")
        validar_nome = False
        break

    if(validar_nome == False):
      continue

    competicao[competidor_atual]['nome'] = nome #O nome válido é adicionado no dicionário do competidor.

    break

  salto_atual = 1 #Variável que iterará sobre os 5 saltos do competidor atual.

  while(salto_atual <= 5):

    salto = float(input(f"Insira a distância do salto {salto_atual}: "))

    if(salto <= 0):
      print("Inválido, insira apenas valores positivos maiores que 0.")
      continue

    competicao[competidor_atual]['saltos'].append(salto) #Registra cada salto pro competidor atual no seu dicionário de dados.

    salto_atual += 1

  salto_atual = 1

  while(True):

    pergunta = input('Outro atleta será adicionado? (S) / (N): ').upper().strip()

    #Validação da pergunta
    if(pergunta == 'S'):

      competidor_atual += 1

      competicao[competidor_atual] = {

            'id': competidor_atual,
            'nome': '',
            'saltos': [],
        }

      break

    elif(pergunta == 'N'):
      break

    else:
      print("Insira apenas entre S(sim) ou N(não) para resposta da pergunta.")
      continue

#Lista de nomenclatura para exibir a nota de cada salto de cada competidor.
nomenclatura_salto = ['Primeiro Salto','Segundo Salto','Terceiro Salto','Quarto Salto','Quinto Salto']

media_atual = 0 #Variável que exibirá a média de salto de cada competidor.

pior_salto = 0 #Variável que exibirá a pior pontuação de salto de cada competidor.

melhor_salto = 0 #Variável que exibirá a melhor pontuação de salto de cada competidor.

resultado = [] #Lista que exibirá a ordem de classificação de acordo com a média de cada competidor.

for j in competicao:
  i = 0
  print(f"Atleta: {competicao[j]['nome']} \n")

  for k in competicao[j]['saltos']:
    print(f"{nomenclatura_salto[i]}: {k} m")
    i += 1

  print('\n')

  pior_salto = min(competicao[j]['saltos']) #Pega e adiciona na variável pior_salto o menor salto do competidor atual.

  melhor_salto = max(competicao[j]['saltos']) #Pega e adiciona na variável melhor_salto o maior salto do competidor atual.

  print(f"Melhor Salto: {melhor_salto}")

  print(f"Pior Salto: {pior_salto}")

  competicao[j]['saltos'].remove(melhor_salto) #Remove o melhor salto da lista de saltos do competidor atual.

  competicao[j]['saltos'].remove(pior_salto) #Remove o pior salto da lista de saltos do competidor atual.

  media_atual = (sum(competicao[j]['saltos']) / 3) #Cálculo da média de cada competidor.

  print(f"Média dos demais saltos: {media_atual:.1f} m")

  print('\n')

  resultado.append((competicao[j]['nome'],media_atual)) #Adiciona na lista resultado o nome e a média de cada competidor.

print("Resultado Final: ")

resultado.sort(key=lambda x: x[1], reverse=True) #O método sort ordenará de modo decrescente com base nas médias a classificação dos
#competidores.

for nome, media in resultado:
  print(f"{nome}: {media:.1f} m")
```

```python
"""
#47) Em uma competição de ginástica, cada atleta recebe votos de sete jurados. A melhor e a pior nota são eliminadas.
A sua nota fica sendo a média dos votos restantes. Você deve fazer um programa que receba o nome do ginasta e as notas dos sete
jurados alcançadas pelo atleta em sua apresentação e depois informe a sua média, conforme a descrição acima informada (retirar o
melhor e o pior salto e depois calcular a média com as notas restantes). As notas não são informados ordenadas. Um exemplo de saída
do programa deve ser conforme o exemplo abaixo:

Atleta: Aparecido Parente
Nota: 9.9
Nota: 7.5
Nota: 9.5
Nota: 8.5
Nota: 9.0
Nota: 8.5
Nota: 9.7

Resultado final:
Atleta: Aparecido Parente
Melhor nota: 9.9
Pior nota: 7.5
Média: 9,04
"""
#Dicionário que armazena os dados dos competidores.
competicao = {
    1: {
        'nome': '',
        'votos': [],
    }
}

competidor_atual = 1 #Variável que iterará sobre cada competidor para fazer seu devido registro no dicionário.

pergunta = '' #Variável que servirá para finalizar o registro dos competidores caso não tenha mais atletas.

while(pergunta != 'N'):

  while(True):

    validar_nome = True #Variável que serve para reiniciar o registro de nome caso seja inválido.

    nome = input(f"Insira o nome do competidor {competidor_atual}: ").capitalize()

    #Validação de nome
    for i in nome:
      if(i.isdigit()):
        print("Inválido, nome não pode ter números.")
        validar_nome = False
        break

    if(validar_nome == False):
      continue

    competicao[competidor_atual]['nome'] = nome #Faz o registro do nome no dicionário do competidor.

    break

  nota_atual = 1 #Variável que itera sobre cada uma das sete notas.

  while(nota_atual <= 7):

    nota = float(input(f"Insira a nota {nota_atual}: "))

    #Validação das notas.
    if(nota < 0):
      print("Inválido, insira apenas valores positivos.")
      continue

    competicao[competidor_atual]['votos'].append(nota) #Registra cada uma das notas na lista do dicionário do competidor atual.

    nota_atual += 1

  while(True):

    pergunta = input("Tem outro competidor para fazer o registro? (S) / (N): ").upper().strip()

    #Validação da pergunta.
    if(pergunta == 'S'):

      competidor_atual += 1

      competicao[competidor_atual] = {
          'nome': '',
          'votos': []
      }

      break

    elif(pergunta == 'N'):
      break

    else:
      print("Inválido, insira apenas (S)sim ou (N)não como resposta.")
      continue

nota_atual = 1 #Essa mesma variável exibirá as notas de cada um dos concorrentes.

for i in competicao:

    print(f"Atleta: {competicao[i]['nome']}")

    for j in competicao[i]['votos']:

      print(f"Nota {nota_atual}: {j}")

      nota_atual += 1

    print('\n')

print("Resultado Final: ")

resultado = [] #Lista onde ficará a ordem de classificação dos participantes.

nome_atleta = '' #Variável que servirá para adicionar o nome do competidor na lista resultado.

melhor_nota = 0 #Variável que servirá para adicionar a melhor nota do competidor na lista resultado.

pior_nota = 0 #Variável que servirá para adicionar a pior nota do competidor na lista resultado.

media = 0 #Variável que servirá para adicionar a média competidor na lista resultado.

for i in competicao:

  nome_atleta = competicao[i]['nome']

  melhor_nota = max(competicao[i]['votos'])

  pior_nota = min(competicao[i]['votos'])

  competicao[i]['votos'].remove(max(competicao[i]['votos']))

  competicao[i]['votos'].remove(min(competicao[i]['votos']))

  media = sum(competicao[i]['votos']) / 5

  resultado.append((nome_atleta, melhor_nota, pior_nota, media))

resultado.sort(key=lambda x: x[3], reverse=True) #Faz a exibição em ordem decrescente com base na média de cada competidor.

for nome, nota_alta, nota_baixa, media in resultado:
  print(f"Atleta: {nome}")
  print(f"Melhor nota: {nota_alta}")
  print(f"Pior nota: {nota_baixa}")
  print(f"Média: {str(f'{media:.2f}').replace('.', ',')}")
  print("\n")
```

```python
"""
#49) Faça um programa que mostre os n termos da Série a seguir:
  S = 1/1 + 2/3 + 3/5 + 4/7 + 5/9 + ... + n/m.
Imprima no final a soma da série.
"""

from fractions import Fraction #Biblioteca para manipular frações.

while(True):

  numerador = 1 #Variável onde ficará o valor dos numeradores.

  denominador = 1 #Variável onde ficará o valor dos denominadores.

  fracoes = [] #Lista onde ficará armazenado as frações, para posterior soma.

  n = int(input("Insira o n-ésimo termo: "))

  #Validação do n-ésimo termo.
  if(n <= 0):
    print("Inválido, insira apenas valores positivos maiores que 0.")
    continue

  iteracao = 1 #Variável que iterará até o n-ésimo termo.

  while(iteracao <= n):

    f = Fraction(numerador, denominador) #O método Fraction transforma os valores numa fração.

    print(f)

    fracoes.append(f)

    numerador += 1

    denominador += 2

    iteracao += 1

  print(f"Soma da série: {sum(fracoes)}")
```
