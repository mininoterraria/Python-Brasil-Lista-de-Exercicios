```python
#03) Faça um Programa que leia 4 notas, mostre as notas e a média na tela.

notas = [] #Lista onde ficará armazenado as notas.

i = 1 #Variável que servirá para iterar e registrar as 4 notas.

while(i <= 4):

  nota = float(input(f"Insira a nota {i}: "))

  if(nota < 0):
    print("Inválido, insira apenas notas positivas.")
    continue
  elif(nota > 10):
    print("Inválido, insira apenas notas entre 0 à 10.")
    continue

  notas.append(nota) #Adiciona a nota na lista.

  i += 1

print(f"Notas: {notas}")

print(f"Média das notas: {sum(notas) / 4:.2f}")
```

```python
#04) Faça um Programa que leia um vetor de 10 caracteres, e diga quantas consoantes foram lidas. Imprima as consoantes.

lista_caracteres = [] #Lista onde ficará os caracteres.

i = 1 #Variável que servirá para iterar e registrar os 10 caracteres.

while(i <= 10):

  caractere = input(f"Insira o caractere {i}: ").upper().strip()

  #Validação do caractere.
  if(len(caractere) > 1):
    print("Inválido, insira apenas 1 único caractere.")
    continue

  elif(caractere.isdigit()):
    print("Inválido, caractere não pode ser numérico.")
    continue

  lista_caracteres.append(caractere)

  i += 1

vogais = ['A','E','I','O','U']

consoantes = 0 #Variável que contabilizará as consoantes.

lista_consoantes = [] #Lista onde ficará apenas as consoantes.

for i in lista_caracteres:
  if(i not in vogais):
    consoantes += 1
    lista_consoantes.append(i)

print(f"Consoantes lidas: {consoantes}")

print(f"Consoantes: {lista_consoantes}")
```

```python
"""
#06) Faça um Programa que peça as quatro notas de 10 alunos, calcule e armazene num vetor a média de cada aluno,
imprima o número de alunos com média maior ou igual a 7.0.
"""

medias_alunos = [] #Lista onde ficará as médias de cada aluno.

numero_alunos = 0 #Variável que ficará armazenado a quantidade de alunos com nota superior ou igual à 7.

for i in range(1,11):

  iterar_nota = 1 #Variável que iterará sobre cada nota.

  media = 0 #Variável que armazenará a média de cada aluno.

  while(iterar_nota <= 4):

    nota = float(input(f"Insira a {iterar_nota}º nota do {i}º aluno(a): "))

    #Validação da nota

    if(nota < 0):
      print("Inválido, nota não pode ser negativa.")
      continue

    elif(nota > 10):
      print("Inválido, nota não pode ser maior que dez.")
      continue

    media += nota

    iterar_nota += 1

  medias_alunos.append(media / 4)

for i in medias_alunos:
  if(i >= 7):
    numero_alunos += 1

print(f"Alunos com notas maiores ou igual a 7: {numero_alunos}")
```

```python
"""
#08) Faça um Programa que peça a idade e a altura de 5 pessoas,
armazene cada informação no seu respectivo vetor. Imprima a idade e a altura na ordem inversa a ordem lida
"""

idades = [] #Lista onde ficará armazenado as idades.

alturas = [] #Lista onde ficará armazenado as alturas.

for i in range(1,6):
  while(True):

    idade = int(input(f"Insira a idade da {i}ª pessoa: "))

    #Validação da idade
    if(idade < 0):
      print("Inválido, insira apenas idades positivas.")
      continue

    idades.append(idade)

    break

  while(True):

    altura = float(input(f"Insira a altura da {i}ª pessoa: "))

    #Validação da altura.
    if(altura <= 0):
      print("Inválido, insira apenas alturas maiores que 0.")
      continue

    alturas.append(altura)

    break

for i in range(4,-1,-1):

  print(f"{i + 1}ª Pessoa: Idade - {idades[i]} / Altura - {alturas[i]}")
```

```python
"""
#12) Foram anotadas as idades e alturas de 30 alunos. Faça um Programa que determine
quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos
"""

idades = [] #Lista onde ficará as idades de cada aluno.

alturas = [] #Lista onde ficará a altura de cada aluno.

i = 1 #Variável que fará as iterações necessárias em cada laço.

while(i <= 5):

  idade = int(input(f"Insira a idade do(a) aluno(a) {i}: "))

  #Validação da idade.
  if(idade < 0):
    print("Inválido, insira apenas valores positivos para idade.")
    continue

  idades.append(idade)

  while(True):

    altura = float(input(f"Insira a altura do(a) aluno(a) {i}: "))

    #Validação da altura.
    if(altura <= 0):
      print("Insira apenas valores positivos maiores que zero para altura.")
      continue

    alturas.append(altura)

    i += 1

    break


media = sum(alturas) / len(alturas) #Cálculo da média de alturas dos alunos.

qtd_alunos = 0 #Variável que contabiliza a quantidade de alunos com mais de 13 anos com altura inferior à média.

iterador = 0 #Variável que fará a iteração pela lista de alturas.

for i in idades:

  if(i > 13 and alturas[iterador] < media):
    qtd_alunos += 1

  iterador += 1

print(f"Média de altura dos alunos: {media:.2f}")

print(f"Quantidade de alunos com mais de 13 anos que possuem altura inferior à média: {qtd_alunos}")
```

```python
"""
#13) Faça um programa que receba a temperatura média de cada mês do ano e armazene-as em uma lista.
Após isto, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual,
e em que mês elas ocorreram (mostrar o mês por extenso: 1 – Janeiro, 2 – Fevereiro, . . . ).
"""

meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro',
         'Dezembro']#Lista armazenando os meses do ano.

temperatura_meses = [] #Lista onde ficará a temperatura média de cada mês.

for i in range(0,12):

  temperatura = float(input(f"Insira a temperatura média do mês de {meses[i]}: "))

  temperatura_meses.append(temperatura)

media_anual = sum(temperatura_meses) / len(temperatura_meses) #Cálculo da média anual da temperatura.

ordem = 1 #Variável que servirá para contar a ordem de cada mês com temperatura maior que a média anual.

print(f"Média anual: {media_anual:.2f} \n")

for i in range(0,12):

  if(temperatura_meses[i] > media_anual):
    print(f"{ordem} - {meses[i]}: {temperatura_meses[i]:.2f} graus.")

  ordem += 1
```

```python
"""
#14) Utilizando listas faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:
"Telefonou para a vítima?"
"Esteve no local do crime?"
"Mora perto da vítima?"
"Devia para a vítima?"
"Já trabalhou com a vítima?" O programa deve no final emitir uma classificação sobre a participação da pessoa
no crime. Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita",
entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".
"""

#Dicionário onde fica armazenada as perguntas que serão feitas.
perguntas = [
    "Telefonou para a vítima?",
    "Esteve no local do crime?",
    "Mora perto da vítima?",
    "Devia para a vítima?",
    "Já trabalhou com a vítima?"
]

i = 0 #Variável que iterará sobre cada pergunta.

classificacao = 0 #Variável que classificará o entrevistado com base nas respostas.

while(i <= 4):

  resposta = input(f"{perguntas[i]}(S) / (N): ").upper().strip()

  #Validação da resposta
  if(resposta == 'S'):
    classificacao += 1

  elif(resposta == 'N'):

    pass
  else:
    print("Apenas dê como resposta (S)sim ou (N)não.")
    continue

  i += 1

#Estrutura que classificará o entrevistado com base nas respostas.
if(classificacao == 2):
  print("Suspeita.")
elif(classificacao == 3 or classificacao == 4):
  print("Cúmplice.")
elif(classificacao == 5):
  print("Assassino.")
else:
  print("Inocente.")
```

```python
"""
#15) Faça um programa que leia um número indeterminado de valores, correspondentes a notas,
encerrando a entrada de dados quando for informado um valor igual a -1 (que não deve ser armazenado).
Após esta entrada de dados, faça:
Mostre a quantidade de valores que foram lidos;
Exiba todos os valores na ordem em que foram informados, um ao lado do outro;
Exiba todos os valores na ordem inversa à que foram informados, um abaixo do outro;
Calcule e mostre a soma dos valores;
Calcule e mostre a média dos valores;
Calcule e mostre a quantidade de valores acima da média calculada;
Calcule e mostre a quantidade de valores abaixo de sete;
Encerre o programa com uma mensagem;
"""

numeros = [] #Lista onde ficará as notas do input.

i = 1 #Variável que iterará sobre cada nota.

qtd_acima_media = 0 #Variável que armazenará a quantidade de valores acima da média.

qtd_abaixo_7 = 0 #Variável que armazenará a quantidade de valores abaixo de 7.

while(True):

  numero = int(input(f"Informe o {i}º valor: "))

  #Validação do número
  if(numero < 0 and numero != -1):
    print("Inválido, o único valor negativo aceito é -1, que encerra o sistema.")
    continue

  elif(numero > 10):
    print("Inválido, insira apenas de 0 à 10 para registrar notas.")
    continue

  elif(numero == -1):
    break

  numeros.append(numero)

  i += 1

#Estrutura para contabilizar a quantidade de valores acima da média.
for i in numeros:
  if(i > (sum(numeros) / len(numeros))):
   qtd_acima_media += 1

#Estrutura para contabilizar a quantidade de valores menor que 7.
for i in numeros:
  if(i < 7):
    qtd_abaixo_7 += 1

print(f"Quantidade de valores lidos: {len(numeros)}")

print(f"Valores em ordem: {numeros}")

print("Valores em ordem inversa: ")

for i in reversed(numeros): #Inverte a lista sem alterar a original.
  print(i)

print(f"Soma dos valores: {sum(numeros)}")

print(f"Média dos valores: {sum(numeros) / len(numeros):.2f}")

print(f"Quantidade de valores acima da média calculada: {qtd_acima_media} ")

print(f"Quantidade de valores abaixo de 7: {qtd_abaixo_7}")

print("Fim do programa!")
```

```python
"""
#16) Utilize uma lista para resolver o problema a seguir. Uma empresa paga seus vendedores com base em comissões. O vendedor recebe $200 por semana mais
9 por cento de suas vendas brutas daquela semana. Por exemplo, um vendedor que teve vendas brutas de $3000 em uma semana recebe $200 mais 9 por cento de
$3000, ou seja, um total de $470. Escreva um programa (usando um array de contadores) que determine quantos vendedores receberam salários nos seguintes

intervalos de valores:
$200 - $299
$300 - $399
$400 - $499
$500 - $599
$600 - $699
$700 - $799
$800 - $899
$900 - $999
$1000 em diante

Desafio: Crie uma fórmula para chegar na posição da lista a partir do salário, sem fazer vários ifs aninhados.
"""

import math #Biblioteca que contém funções de arredondamento.


intervalos = [ #Lista armazenando a nomenclatura dos intervalos.
    '$200 - $299',
    '$300 - $399',
    '$400 - $499',
    '$500 - $599',
    '$600 - $699',
    '$700 - $799',
    '$800 - $899',
    '$900 - $999',
    '$1000 em diante'

]

contador_intervalo = [0,0,0,0,0,0,0,0,0] #Lista onde fica a quantidade de cada intervalo de valores. indice 0 à 8.

while(True):

  venda = float(input("Insira o valor da venda: "))

  #Validação de venda.
  if(venda < 0):
    print("Inválido, insira apenas valores positivos.")
    continue

  #Encerra o sistema mesmo tendo ou não registros de venda.
  if(venda == 0):
    break

  salario_bruto = 200 #Variável que armazena o valor fixo de 200 que os vendedores recebem.

  comissao = venda * 0.09 #Comissão que o vendedor recebe do seu total de vendas.

  salario_liquido = salario_bruto + comissao #Salário total que o vendedor recebe do seu fixo + 9% de comissão.

  intervalo_valores = math.ceil((salario_liquido - salario_bruto) / 100) #Cálculo para definir o
  #intervalo de valores com base na venda do vendedor.

  print(intervalo_valores)

  #Caso no cálculo dê um valor muito além do range da lista, é redirecionado para o último.

  if(intervalo_valores > 9):
    intervalo_valores = 9

  contador_intervalo[intervalo_valores - 1] += 1 #Incrementa o intervalo certo.

for i in range(0,9):

  print(f"Vendedores que receberam salário no intervalo {intervalos[i]}: {contador_intervalo[i]}")
```

```python
'''
#17) Em uma competição de salto em distância cada atleta tem direito a cinco saltos.
O resultado do atleta será determinado pela média dos cinco valores restantes. Você deve fazer um programa que
receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe o nome, os saltos e a
média dos saltos. O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa
deve ser conforme o exemplo abaixo:
Atleta: Rodrigo Curvêllo

Primeiro Salto: 6.5 m
Segundo Salto: 6.1 m
Terceiro Salto: 6.2 m
Quarto Salto: 5.4 m
Quinto Salto: 5.3 m

Resultado final:
Atleta: Rodrigo Curvêllo
Saltos: 6.5 - 6.1 - 6.2 - 5.4 - 5.3
Média dos saltos: 5.9 m
'''

#Dicionário onde ficará os dados de cada atleta.
competicao = {

}

#Lista que armazena a ordem de cada salto, será usada para nomenclatura.
ordem_saltos = ['Primeiro Salto','Segundo Salto','Terceiro Salto','Quarto Salto','Quinto Salto']

iterar_nome = 1 #Variável que iterará sobre cada atleta

while(True):

  nome_atleta = input(f"Informe o nome do(a) {iterar_nome}º atleta: ").strip().title()

  #Estrutura de decisão que impede de encerrar o programa antes de registrar ao menos 1 atleta.
  if(not competicao and nome_atleta == ''):
    print("Inválido, registre ao menos 1 atleta antes de encerrar o sistema.")
    continue

  #Se for preenchido um campo vazio, encerra o programa.
  if(nome_atleta == ''):
    break

  else:

    #Caso seja preenchido um nome, será colocado os dados do atleta no dicionário.
    competicao[iterar_nome] = {
        'nome': '',
        'saltos': [],
        'media_saltos': 0
    }

    competicao[iterar_nome]['nome'] = nome_atleta #Registra o nome do atleta no dicionário.

    iterar_salto = 1 #Variável que iterará sobre cada salto.

    while(iterar_salto <= 5):

      salto = float(input(f"Insira a nota do {iterar_salto}º salto do(a) {iterar_nome}º atleta: "))

      #Validação de salto
      if(salto <= 0):
        print("Inválido, insira apenas valores positivos maiores que 0.")
        continue

      competicao[iterar_nome]['media_saltos'] += salto #Registra cada salto na lista de média para posterior cálculo.

      competicao[iterar_nome]['saltos'].append(f"{salto:.1f}") #Registra o salto atual do atleta atual na sua lista.

      iterar_salto += 1

    iterar_nome += 1

for i in competicao:

  iterar_saltos = 0 #Variável que iterará sobre cada salto para exibição.

  print(f"Atleta: {competicao[i]['nome']} \n")

  for j in ordem_saltos:

    print(f"{j}: {competicao[i]['saltos'][iterar_saltos]} m")

    iterar_saltos += 1

  competicao[i]['media_saltos'] = competicao[i]['media_saltos'] / len(competicao[i]['saltos']) #Cálculo da média

  competicao[i]['saltos'] = ' - '.join(map(str,competicao[i]['saltos'])) #Formata a exibição dos saltos separando-os
  #por hífen

  print('\n')

print("Resultado Final: \n")

for i in competicao:

  print(f"Atleta: {competicao[i]['nome']}")

  print(f"Saltos: {competicao[i]['saltos']}")

  print(f"Média dos saltos: {competicao[i]['media_saltos']:.1f} m\n")
```
